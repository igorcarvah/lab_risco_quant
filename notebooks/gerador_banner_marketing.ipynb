```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
import os

os.makedirs('assets', exist_ok=True)

# 1. Configuração Dark Mode (Estilo Terminal)
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(3, 3), dpi=150) # Aumentei dpi para ficar HD
fig.patch.set_facecolor('#0e1117') # Cor de fundo chique (tipo VS Code)
ax.set_facecolor('#0e1117')

# 2. Dados
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

# 3. Desenha a Curva Principal (Ciano Neon)
ax.plot(x, y, color='#00d4ff', lw=3)
ax.fill_between(x, y, color='#00d4ff', alpha=0.1)

# 4. O Pulo do Gato: Pintar o Risco (A Cauda Esquerda)
# Vamos simular um VaR de 5% (abaixo de -1.64 desvios)
x_risk = np.linspace(-4, -1.64, 100)
y_risk = norm.pdf(x_risk, 0, 1)
ax.fill_between(x_risk, y_risk, color='#ff2b2b', alpha=0.8) # Vermelho Sangue

# 5. Limpeza Total
ax.axis('off')
ax.grid(False)

# 6. Salvar
plt.tight_layout()
plt.savefig('assets/logo_lab_risco_v2.png', facecolor='#0e1117')
print("✅ Logo V2 (Dark Mode) gerado: assets/logo_lab_risco_v2.png")
```
